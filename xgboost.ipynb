```python
import xgboost as xgb
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
```

```python
from numpy.random import random_sample
x,y=make_classification(n_samples=5000,n_features=10,random_state=42)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.utils import random
model=xgb.XGBClassifier(n_estimaotrs=100,max_depth=4,learning_rate=0.1,random_state=42)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
print(f"classification report is {classification_report(y_test,y_pred)}")
```
